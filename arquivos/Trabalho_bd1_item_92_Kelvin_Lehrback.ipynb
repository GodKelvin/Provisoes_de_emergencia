**9.9 CONSULTAS COM SELF JOIN E VIEW (Mínimo 6)**

Não é possível realizar uma consulta com self join, pois neste modelo de banco de dados não há tabelas cujo o valor de uma coluna seja outra coluna dessa mesma tabela.

*`Criando um cursor`*

```
cursor_elephant = None
cursor_elephant = conn.cursor()
```

```
cursor_elephant.execute(""" CREATE or REPLACE VIEW pessoa_consome_produto AS
                            SELECT pessoa.nome_pessoa, produto.nome_produto AS consome
                            FROM pessoa
                            INNER JOIN pessoa_consome
                            ON(pessoa_consome.fk_cod_pessoa = pessoa.cod_pessoa)
                            INNER JOIN produto
                            ON(produto.cod_produto = pessoa_consome.fk_cod_produto)
                            ORDER BY nome_pessoa""", conn)
```
